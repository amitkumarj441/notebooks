```python
import matplotlib as mpl
mpl.rcParams['mathtext.fontset'] = 'cm'

import numpy as np
import plotnine as p9
import scipy as sp
import pandas as pd

x = np.linspace(-1, 1, 100)
y_levy = sp.stats.levy.pdf(x, loc=-1)
y_laplace = sp.stats.laplace.pdf(x, loc=-0.3)
y_wigner = sp.stats.semicircular.pdf(x, scale=1.25, loc=0.5)
y_tsine = np.abs(np.sin(3 * np.pi * x)) / 4

tex_ks = r'$\dfrac{\sqrt{2\pi}}{x} \sum_{k=1}^\infty \ \exp\left( \dfrac{-(2k - 1)^2 \pi^2} {8x^2} \right)$'
tex_levy = r'$\sqrt{\dfrac{\gamma}{2\pi}} \ \dfrac{\exp\left( \dfrac{-\gamma}{2(x - \mu)} \right)}{(x - \mu)^{3/2}}$'
tex_cauchy = r'$\dfrac{1}{\pi} \arctan\left( \dfrac{x - \mu}{\gamma} \right) + \dfrac{1}{2}$'

p9.ggplot(pd.DataFrame(dict(x=x)), p9.aes(x='x', y='y')) \
    + p9.geom_area(data=pd.DataFrame(dict(y=y_levy)), fill='red', alpha=0.2) \
    + p9.geom_area(data=pd.DataFrame(dict(y=y_laplace)), fill='green', alpha=0.2) \
    + p9.geom_area(data=pd.DataFrame(dict(y=y_wigner)), fill='yellow', alpha=0.2) \
    + p9.geom_area(data=pd.DataFrame(dict(y=y_tsine)), fill='blue', alpha=0.2) \
    + p9.annotate('text', label=r'Memes', x=0, y=0.5, size=20) \
    + p9.annotate('text', label=r'and you choose to be normal', x=0, y=0.04, size=20) \
    + p9.annotate('text', label=tex_ks, x=0, y=0.4, alpha=0.3, size=15) \
    + p9.annotate('text', label=tex_levy, x=0.4, y=0.275, alpha=0.3, size=15) \
    + p9.annotate('text', label=tex_cauchy, x=-0.2, y=0.15, alpha=0.3, size=15) \
    + p9.theme(axis_text=p9.element_blank(), axis_ticks=p9.element_blank()) \
    + p9.labs(x=None, y=None)
```
